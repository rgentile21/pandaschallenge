```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
purchase_path = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(purchase_path, encoding="utf-8")
purchase_df.head()
```

## Player Count

* Display the total number of players


```python
#purchase_df = pandas.read_csv('purchase_data.csv', header=None)
sntotal = purchase_df.drop_duplicates(['SN'], keep='first')

#Create a Data Frame for Player Count
player_count = len(purchase_df["SN"].unique())
player_count = pd.DataFrame({"Total Players":[player_count]})
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#calculate the number of unique items
item_count = len(purchase_df["Item Name"].unique())

#Average Purchase Price
average_price = purchase_df["Price"].mean()

#Number of Items Purchased
purchase_num = len(purchase_df["Item Name"])

#Total Revenue
total_revenue = purchase_df["Price"].sum()
```

```python
#Create New Data Frame - Summary Data Frame for Purchasing Analysis Total
Purchasing_Analysis = pd.DataFrame({
                              "Unique Items": [item_count],
                              "Average Price": [average_price],
                              "Number of Purchases": [purchase_num],
                              "Total Revenue": [total_revenue]})
#Data Frame Formatting
Purchasing_Analysis = Purchasing_Analysis[["Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]
Purchasing_Analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Gender Demographics
#Count of Gender - Reviewed with Learning Assistant
genderttl = sntotal['Gender'].value_counts()

#Gender Demographic DataFrame
genderttl = pd.DataFrame(genderttl)

#Percentage of Players
genderttl['Percentage of Players'] = genderttl['Gender'] / len(sntotal)

#Format Columns
genderttl = genderttl[['Percentage of Players', 'Gender']]
genderttl = genderttl.rename(columns={'Gender':'Total Count'})

#Percentage
genderttl['Percentage of Players'] = genderttl['Percentage of Players'].map('{:.2%}'.format)

genderttl


                              
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Groupby Gender
genderprc = purchase_df.groupby(['Gender'])

#Purchase Count
genderprctl = genderprc['Price'].count()

#Average Price of Purchase
genderavprc = genderprc['Price'].mean()

#Total Purchase Price
genderpurprc = genderprc['Price'].sum()

#Revision of Total Number of Users (Per Review with Learning Assistant)
trueplayers = genderpurprc/genderttl['Total Count']



#DataFrame for Purchasing Analysis by Gender

genderprc = pd.DataFrame({
                              "Purchase Count": genderprctl,
                              "Average Purchase Price": genderavprc,
                              "Total Purchase Value": genderpurprc,
                              "Average Purchase Total": trueplayers})
                              
genderprc
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Establish Bins for Ages and Categorize Players Using Age Bins
bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]

#Group Names
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
               
#Cut Data Set Per Hint
sntotal['Age Group'] = pd.cut(sntotal.loc[:,"Age"], bins, labels=group_names)

#Group By Age Group
agegroup = sntotal.groupby("Age Group")

#Create DataFrame for Age 
agegroupval = agegroup['SN'].count()

#Create DataFrame for Age Group
agegroup = pd.DataFrame({'Player Count':agegroupval})

#Percentage Calculation of Players
agegroup['Percentage of Players'] = agegroup['Player Count']/len(sntotal)

#Output (Index per Panda Documentation Review)
agegroup.index.names = ['Age']
agegroup['Percentage of Players'] = agegroup['Percentage of Players'].map('{:.2%}'.format)

#Show Results
agegroup
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Establish Purchase Analysis by Age
agepurchase = sntotal.groupby("Age Group")

#Purchase Count by Age
agepurchcnt = agepurchase['Price'].count()

#Average Price of Purchase by Age
agepurcavg = agepurchase['Price'].mean()

#Total Purchase Price by Age
agepurcttl = agepurchase['Price'].sum()

#Revision of Total Number of Users by Age
ageplyrpur = agepurcttl/agegroup['Player Count']

#DataFrame for Purchasing Analysis by Age

agepurchase = pd.DataFrame({
                              "Purchase Count": agepurchcnt,
                              "Average Purchase Price": agepurcavg,
                              "Total Purchase Value": agepurcttl,
                              "Average Purchase Total": ageplyrpur})
                              
agepurchase


```

## Most Popular Items

* Create a summary data frame to hold the results


* Sort the purchase count column in descending order

```python
#Retrieve Item ID, Item Name, Item Price
mostpopitem = purchase_df.groupby(['Item ID','Item Name'])

#Most Popular Items
popitemcnt = mostpopitem['Item ID'].count()

#Average Price of Purchase for Top Spenders
popitemavg = mostpopitem['Price'].mean()

#Total Purchase Price by Age
popitemttl = mostpopitem['Price'].sum()

mostpopitem = pd.DataFrame({
                              "Purchase Count": popitemcnt,
                              "Average Purchase Price": popitemavg,
                              "Total Purchase Value": popitemttl})

#Ascending by Total Purchase Value
mostpopitem = mostpopitem.sort_values(['Purchase Count'],ascending=False)

mostpopitem.head(10)
```
